```python
# Point to the datasets stored on the cloud
ratings_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-ML321EN-SkillsNetwork/labs/datasets/ratings.csv"
ratings_df = pd.read_csv(ratings_url)
```

```python
from surprise import KNNBasic
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split, KFold
from surprise import accuracy
```

```python
ratings_df.to_csv("course_ratings.csv", index=False)
# Read the course rating dataset with columns user item rating
reader = Reader(
        line_format='user item rating', sep=',', skip_lines=1, rating_scale=(2, 3))

#coruse_dataset = Dataset.load_from_file("course_ratings.csv", reader=reader)
coruse_dataset = Dataset.load_from_df(ratings_df[["user", "item", "rating"]], reader)
```

```python
trainset, testset = train_test_split(coruse_dataset, test_size=.3)
```

```python
print(f"Total {trainset.n_users} users and {trainset.n_items} items in the trainingset")
```

```python
kf = KFold(n_splits=3)

algo = KNNBasic() 
# need to play with multiple params on this model to see which params are 
# effecient, as well need to apply GridSearchCV() to indetify some of these
# better hyper parameter values

for trainset, testset in kf.split(coruse_dataset):

    # train and test algorithm.
    algo.fit(trainset)
    predictions = algo.test(testset)

    # Compute and print Root Mean Squared Error
    accuracy.rmse(predictions, verbose=True)
```
